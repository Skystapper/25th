```python
#@markdown #STEP 2️⃣: Perlengkapan Models Downloader
from IPython.display import clear_output
import os
import requests
from urllib.parse import unquote

def download_model(Type, DirectLink_URL):
    if Type == "None" or not DirectLink_URL:
        print("Skipping download as no valid Type or DirectLink_URL provided.")
        return

    output_path = "/content/Fooocus/models/"

    if Type == "Checkpoint":
        output_path += "checkpoints/"
    elif Type == "LoRA":
        output_path += "loras/"
    elif Type == "Textual Inversion":
        output_path += "embeddings/"
    else:
        print("Invalid type specified.")
        return

    # Create the directories if they don't exist
    os.makedirs(output_path, exist_ok=True)

    urls = [url.strip() for url in DirectLink_URL.split(",")]

    for url in urls:
        response = requests.get(url, stream=True)

        if response.status_code == 200:
            content_disposition = response.headers.get('content-disposition')
            if content_disposition:
                filename = unquote(content_disposition.split('filename=')[1])
            else:
                filename = unquote(url.split("/")[-1])  # Extracting the filename from the URL

            # Remove double quotes and semicolons from the filename
            filename = filename.replace('"', '').replace(';', '')

            filename = os.path.join(output_path, filename)  # Modify the filename to include the output path

            if not os.path.exists(filename):
                print("Downloading file:", filename)
                chunk_size = 5242880  # 5 MB
                with open(filename, 'wb') as f:
                    for chunk in response.iter_content(chunk_size=chunk_size):
                        if chunk:
                            f.write(chunk)
                print("File downloaded successfully.")
            else:
                print("File already exists:", filename)
        else:
            print("Failed to download the file:", url)

    # Clear the output to keep the notebook clean
    clear_output()

    # Print the success message
    print("✅ Done", "success")

# Call the function with the first set of parameters
Type1 = "Checkpoint"  # @param ["Checkpoint", "LoRA", "Textual Inversion"]
DirectLink_URL1 = "https://civitai.com/api/download/models/464939"  # @param {'type': 'string'}
download_model(Type1, DirectLink_URL1)

# Call the function with the second set of parameters
Type2 = "Checkpoint"  # @param ["None", "Checkpoint", "LoRA", "Textual Inversion"]
DirectLink_URL2 = "https://civitai.com/api/download/models/290640?type=Model&format=SafeTensor&size=pruned&fp=fp16"  # @param {'type': 'string'}
download_model(Type2, DirectLink_URL2)

# Call the function with the second set of parameters
Type3 = "Checkpoint"  # @param ["None", "Checkpoint", "LoRA", "Textual Inversion"]
DirectLink_URL3 = "https://civitai.com/api/download/models/290640?type=Model&format=SafeTensor&size=pruned&fp=fp16"  # @param {'type': 'string'}
download_model(Type3, DirectLink_URL3)

# Call the function with the second set of parameters
Type4 = "Checkpoint"  # @param ["None", "Checkpoint", "LoRA", "Textual Inversion"]
DirectLink_URL4 = "https://civitai.com/api/download/models/290640?type=Model&format=SafeTensor&size=pruned&fp=fp16"  # @param {'type': 'string'}
download_model(Type4, DirectLink_URL4)

# Call the function with the second set of parameters
Type5 = "Checkpoint"  # @param ["None", "Checkpoint", "LoRA", "Textual Inversion"]
DirectLink_URL5 = "https://civitai.com/api/download/models/290640?type=Model&format=SafeTensor&size=pruned&fp=fp16"  # @param {'type': 'string'}
download_model(Type5, DirectLink_URL5)

# Call the function with the second set of parameters
Type6 = "Checkpoint"  # @param ["None", "Checkpoint", "LoRA", "Textual Inversion"]
DirectLink_URL6 = "https://civitai.com/api/download/models/290640?type=Model&format=SafeTensor&size=pruned&fp=fp16"  # @param {'type': 'string'}
download_model(Type6, DirectLink_URL6)
```
